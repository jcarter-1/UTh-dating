# Notes
* In both cases, there is an assumption that the initial thorium is normally distributed meaning it has a mean and symmetric uncertainty. This may not be true in all cases. The IBIS appraoch (Kinsley et al. 2025 in prep) uses a non-parametric distribution for the global speleothem which more likely relfects this parameter.
* In such a case the traditional linear approximation to uncertainty should not be used. Monte Carlo should be used to capture the potential asymmtery in the initial thorium distribution. 

# Ages
* Following correction the use can then get the ages and uncertainties of the sample using the estimated initial thorium correction
* This is done using the Use_Correction_for_Ages function in the class.
* An example is shown below
* The function returns the age and the age uncertainty (1$
  \sigma$)
* Uncertainties are calculated using linear matrix approximation.


# Notes on ages
* This algorithm uses a tolerance adjustment if no solution can be found. No solution means that there is no value and uncertainty for the initial thorium composition that can align the ages in stratigraphic order.
* In some cases, some ages will not be in strat order. It is then upto the user if then treat this as an outlier or use a more complex and flexible method like the Integrated Baeysian Initial thorium age-depth model for speleothems (IBIS) like presented in Kinsley et al. (2025) in prep.
